```python
fairy_dict = {'Name':['Clefairy','Wigglytuff','Jigglypuff','Igglybuff','Togepi','Marill','Snubbull','Azurill','Azumarill','Cottonee','Whimsicott','Primarina','Morelull','Shiinotic','Tapu Bulu','Tapu Fini',
                      'Rapidash','Mr.Mime','Granbull','Kirlia','Mawile','Spritzee','Dedenne','Carbink','Klefki','Xerneas','Rimnbobee','Mimikyu','Tapu Koko','Tapu Lele','Magearna',
                      'Hatterene','Fidough','Daschbun','Tinkatink','Tinkatuff','Tinkaton','Scream Tail','Flutter Mane'],
           'HP':[70,140,115,90,35,70,60,50,100,40,60,80,40,60,70,70,65,40,90,38,50,78,67,50,57,126,60,55,70,70,80,57,37,57,50,65,85,115,55],
           'Damage':[45,70,45,30,20,20,80,20,50,27,67,74,35,45,130,75,100,45,120,35,85,52,58,50,80,131,55,90,115,85,95,90,55,80,45,55,75,65,55],
           'Type':['Normal','Normal','Normal','Normal','Normal','Water','Normal','Normal','Water','Grass','Grass','Water','Grass','Grass','Grass','Water','Psychic','Psychic','Normal','Psychic','Steel','Normal','Electric',
                   'Rock','Steel','Normal','Bug','Ghost','Electric','Psychic','Steel','Psychic','Normal','Normal','Steel','Steel','Steel','Psychic','Ghost']}
```

Assigning to a variable:

```python
pokemon = pd.DataFrame.from_dict(fairy_dict)
pokemon
```

Displaying the health to damage ratio as a graph

```python
sns.lineplot(data=pokemon,x="HP",y="Damage")
```

Data as a scatterplot by types associated with hp and damage

```python
sns.scatterplot(data=pokemon, x="HP",y="Damage",hue="Type",palette="pastel")
```
